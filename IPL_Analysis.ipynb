But the above seems too vague! Right?

let's go for season wise man of the match using while loop

```python

x, y = 2008, 2017
color = ['Grey', 'Olive', 'Brown', 'Purple','Navy', 'Green', 'Yellow', 'Orange', 'Tomato']
while x < y:
    top_players = ipl[ipl['season'] == x]['player_of_match'].value_counts().head(5)
    plt.bar(list(top_players.index), list(top_players), color= color[x-2008])
    plt.title("Top 5 'Man of the Match' - Season " + str(x))
    plt.xlabel('Players')
    plt.ylabel('No of Matches')
    plt.show()
    
    x += 1

```

Analysis based on the batting first teams

```python
batting_first = ipl[ ipl['win_by_runs'] != 0]

batting_first.head()
```

Plotting Histogram

```python
plt.figure(figsize = (7,7))

plt.hist(batting_first['win_by_runs'], color ='Indigo')
plt.title('Distribution of Runs')
plt.xlabel("Runs")
plt.ylabel("No of Matches Won")

plt.show()
```

Here in the Histogram, we can see that most of the teams have won the match by 20 runs whereas very few teams have won the match by 140 Runs.

```python
batting_first['winner'].value_counts()
```

From the above data, we can see that out of the batting first teams, Chennai Super Kings have won the most matches.

The Bar Plot for the above data is below

```python
plt.figure(figsize = (7,7))
plt.bar(list(batting_first['winner'].value_counts().keys()[0:4]), list(batting_first['winner'].value_counts()[0:4]), color =["yellow", "blue", "purple", "Red"])
plt.title("Top 4 Winning Teams after first batting")
plt.xlabel("Winning Teams")
plt.ylabel('Matches Won')
plt.show()
```

The Pie Chart below, shows the frequency of winning after batting first

```python
plt.figure(figsize = (7,7))
plt.pie( list(batting_first['winner'].value_counts()), labels = list(batting_first['winner'].value_counts().keys()), autopct = '%0.1f%%')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have loaded the data into a dataframe called 'ipl'

# Create a new column indicating whether the toss winner is the same as the match winner
ipl['toss_match_same'] = ipl['toss_winner'] == ipl['winner']

# Create the countplot
sns.countplot(x='toss_match_same', data=ipl)

# Set x-axis labels
plt.xticks([0, 1], ['Different', 'Same'])

# Set plot title and labels
plt.title("Toss Winner vs Match Winner")
plt.xlabel("Toss Winner Same as Match Winner")
plt.ylabel("Count")

# Show the plot
plt.show()
```
